```python
import polars as pl

# Creating a sample DataFrame with different data types
df = pl.DataFrame({
    "Amount": [12345.6789, 23456.7890, 34567.8901],
    "Description": ["Payment for services", "Refund", "Salary"],
    "Date": [pl.date(2024,1,1), pl.date(2024,1,2), pl.date(2024,1,3)],
    "Items": [["apple", "banana"], ["orange", "grape"], ["mango", "kiwi"]]
})

# Display the DataFrame without any custom configurations
print("DataFrame without custom configurations:")
print(df)
```

```python
pl.Config.set_decimal_separator(",")  # Use comma as the decimal separator
pl.Config.set_float_precision(1)  # Display two decimal places for floating-point values
pl.Config.set_fmt_str_lengths(10)  # Limit string display length to 15 characters
pl.Config.set_thousands_separator("_")  # Use comma as the thousands separator
pl.Config.set_trim_decimal_zeros(True)  # Remove trailing zeros from decimals

# Display the DataFrame with formatting applied
print("\nDataFrame with custom configurations:")
print(df)
```
