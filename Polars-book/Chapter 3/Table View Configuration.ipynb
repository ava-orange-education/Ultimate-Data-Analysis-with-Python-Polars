```python
import polars as pl

# Creating a sample DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Salary": [50000.0, 60000.0, 70000.0]
}

df = pl.DataFrame(data)

# Display the DataFrame without any custom configurations
print("DataFrame without custom configurations:")
print(df)
```

```python
# Configuring Polars to display the DataFrame with custom settings
pl.Config.set_ascii_tables(True)  # Use ASCII characters for table outlines

pl.Config.set_tbl_cols(2)  # Display only two columns
pl.Config.set_tbl_rows(2)  # Display only two rows
pl.Config.set_tbl_formatting("MARKDOWN")  # Use Markdown formatting
pl.Config.set_tbl_column_data_type_inline(True)  # Display data type inline

# Display the DataFrame with custom configurations
print("\nDataFrame with custom configurations:")
print(df)
```
