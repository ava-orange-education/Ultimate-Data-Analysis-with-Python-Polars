```python
import polars as pl

# Create three Series
series1 = pl.Series("ages", [25, 30, 35, 40])            # Integer series
series2 = pl.Series("heights", [5.8, 6.0, 5.5, 5.9])     # Float series
series3 = pl.Series("names", ["Alice", "Bob", "Charlie", "David"])  # String series

# Create a DataFrame using the Series
df = pl.DataFrame([series1, series2, series3])

# Print the resulting DataFrame
print(df)
```
