```python
import polars as pl

# Sample DataFrame
df = pl.DataFrame({
    "Values": [1, 2, 3, 4, 5]
})

# Without verbose logging
df_sum = df.select(pl.col("Values").sum())

print("Without verbose logging:")
print(df_sum)
```

```python
import polars as pl

# Enable verbose logging
pl.Config.set_verbose(True)

# Sample DataFrame
df = pl.DataFrame({
    "Values": [1, 2, 3, 4, 5]
})

# With verbose logging
df_sum = df.select(pl.col("Values").sum())

print("\nWith verbose logging:")
print(df_sum)
```
