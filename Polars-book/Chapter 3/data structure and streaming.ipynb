```python
import polars as pl

# Sample DataFrame
df_without = pl.DataFrame({
    "Values": [1, 2, 3, 4, 5],
    "Values2": [1, 2, 3, 4, 5]
})

# Multi-output expressions without auto structify
df_without = df_without.select(pl.all())

print("Without Auto Structify:")
print(df_without)
```

```python
import polars as pl

# Enable auto structify
pl.Config.set_auto_structify(True)

# Sample DataFrame


# Multi-output expressions with auto structify
df_with = df_without.select(pl.all())
print("\nWith Auto Structify:")
print(df_with)
```
