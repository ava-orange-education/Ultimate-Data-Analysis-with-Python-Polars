```python
import polars as pl
import sys

# Define a Config instance for markdown formatting
cfg_markdown = pl.Config(tbl_formatting="MARKDOWN", apply_on_context_enter=True)

# Example function to display patient data in markdown format
@cfg_markdown
def display_patient_report(df: pl.DataFrame) -> None:
    sys.stdout.write(str(df))

# Sample healthcare data
patient_data = pl.DataFrame({
    "Patient ID": [101, 102, 103],
    "Name": ["John Doe", "Jane Smith", "Jim Brown"],
    "Age": [45, 50, 60],
    "Diagnosis": ["Hypertension", "Diabetes", "Asthma"]
})

# Call the function with the applied markdown formatting
display_patient_report(patient_data)
```

```python
# Define Config instances for verbose logging and markdown formatting
cfg_verbose = pl.Config(verbose=True, apply_on_context_enter=True)
cfg_markdown = pl.Config(tbl_formatting="MARKDOWN", apply_on_context_enter=True)

# Example function for detailed data processing with verbose logging
@cfg_verbose
def process_patient_data(df: pl.DataFrame) -> None:
    # Simulate processing the patient data with verbose logging
    df.select(pl.col("Age").sum())
    print("Processed patient data")

# Example function for displaying patient report with markdown formatting
@cfg_markdown
def generate_patient_report(df: pl.DataFrame) -> None:
    sys.stdout.write(str(df))

# Sample healthcare data
patient_data = pl.DataFrame({
    "Patient ID": [101, 102, 103],
    "Name": ["John Doe", "Jane Smith", "Jim Brown"],
    "Age": [45, 50, 60],
    "Diagnosis": ["Hypertension", "Diabetes", "Asthma"]
})

# Call the functions with appropriate configurations
process_patient_data(patient_data)
generate_patient_report(patient_data)
```
